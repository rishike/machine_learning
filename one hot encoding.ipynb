One HotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
counts
```

```python
threshold = 100
df['brand'].nunique()
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'))
```

```python
df['brand'] = df['brand'].apply(lambda x: 'uncommon' if counts[x] < 100 else x)
```
